### 03) WAP to demonstrate the use of Bar chart.

```python
import matplotlib.pyplot as plt
%matplotlib inline
x = [1,2,3,4,5]
y = [5.9,6.2,3.2,8.9,9.7]
l = ['Grade-A','Grade-B','Grade-C','Grade-D','Grade-E']
c = ['b','r','g','y','m']
w = [0.5,0.6,0.3,0.8,0.9]
plt.title("Grade Sheet")
plt.bar(x,y,color=c,label=l,width=w)
plt.show()
```

### 01) WAP to display the value of each bar in a bar chart using Matplotlib.

```python
import matplotlib.pyplot as plt
def addlabels(x,y):
	for i in range(len(x)):
		plt.text(i,y[i],y[i])

if __name__ == '__main__':
	x = ["Engineering", "Hotel Management", "MBA",
		"Mass Communication"]
	y = [9330, 4050, 3030, 5500]
	
	plt.bar(x, y)
	addlabels(x, y)
	plt.title("College Admission")
	
	plt.xlabel("Courses")
	plt.ylabel("Number of Admissions")
	
	plt.show()
```
